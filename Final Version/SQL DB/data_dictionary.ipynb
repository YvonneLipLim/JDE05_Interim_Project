### Table Definition

| Table name      | Table description                                                                                                                                                                 |
| --------------- | --------------------------------------------------------------------------------------------------------------------------------------------------------------------------------- |
| air_temp        | Storing air temperature date time and readings                                                                                                                                    |
| humidity        | Storing humidity date time and readings                                                                                                                                           |
| station         | Normalized and newly created dimension table to store air temperature and humidity station-related information. Attribute includes: station_name, device_id, latitude, longitude |
| locations       | Newly created table to map weather station location with HDB flat location                                                                                                       |
| resale_flat_txn | Storing resale flat price information
