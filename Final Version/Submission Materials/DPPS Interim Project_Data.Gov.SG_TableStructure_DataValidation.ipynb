### Database Schema
We chose a relational model over a dimensional galaxy model for our interim project for the following reasons:

**Simplicity of Implementation and Team Efficiency:** With three fact tables, relational model is simpler to implement compared to dimensional galaxy model, making it manageable and realistic for tight deadlines and also beginners. Additionally, a relational model reduces the need for complex joins in analysis phase.

**Next Step**<br />
Nevertheless, we recognize denormalization simplifies certain queries and improves read performance, but can introduce insert, update and delete anomalies, causing forced insertion of redundant or incomplete data, and also risk of data inconsistencies.
Moving forward, in order to reduce redundancy and enhancing data integrity, we would continue to explore on dimensional galaxy model.

## Table Structure

#### TABLE: air_temp
| Column Name  | Description                             | Data Type | Constraint            |
| ------------ | --------------------------------------- | --------- | --------------------- |
| airtemp_id   | id for air temperature                  | int4      | Primary Key, NOT NULL |
| station_id   | id for weather station                  | varchar   | FOREIGN KEY, NOT NULL |
| temperature  | Air temperature in degree celcius       | numeric   | NOT NULL              |
| airtemp_date | Timestamp (per hour) of air temperature | timestamp | NOT NULL              |

#### TABLE: humidity
| Column Name       | Description                              | Data Type | Constraint            |
| ----------------- | ---------------------------------------- | --------- | --------------------- |
| humidity_id       | id for humidity reading                  | int4      | Primary Key, NOT NULL |
| station_id        | id for weather station                   | varchar   | FOREIGN KEY, NOT NULL |
| humidity_date     | Timestamp (per hour) of humidity reading | timestamp | NOT NULL              |
| humidity_readings | Relative humidity (%)                    | numeric   | NOT NULL              |

#### TABLE: resale_flat_txn
| Column Name         | Description                               | Data Type | Constraint            |
| ------------------- | ----------------------------------------- | --------- | --------------------- |
| resale_id           | id for resale flat                        | int4      | Primary Key, NOT NULL |
| resale_date         | transaction date (YYYY-MM) of resale flat | date      | NOT NULL              |
| town_name           | town name of resale flat                  | varchar   | FOREIGN KEY, NOT NULL |
| flat_type           | flat type of resale flat                  | varchar   | NOT NULL              |
| block_no            | block number of resale flat               | varchar   | NOT NULL              |
| street_name         | street name of resale flat                | varchar   | NOT NULL              |
| storey_range        | storey range of resale flat               | varchar   | NOT NULL              |
| floor_area_sqm      | floor area in square meter                | float8    | NOT NULL              |
| flat_model          | flat model of resale flat                 | varchar   | NOT NULL              |
| lease_commence_year | lease commence year                       | int4      | NOT NULL              |
| remaining_lease     | remaining lease (years and months)        | varchar   | NOT NULL              |
| resale_price        | pricing of resale flat                    | float8    | NOT NULL              |

#### TABLE: station
| Column Name  | Description                  | Data Type | Constraint            |

| latitude     | latitude of weather station  | numeric   | NOT NULL              |
| longitude    | longitude of weather station | numeric   | NOT NULL              |

#### TABLE: locations
| Column Name | Description              | Data Type | Constraint            |
| ----------- | ------------------------ | --------- | --------------------- |

## Data Validation
Generally, the following data validations are applied to validate input, ensure data integrity, and maintains consistency across the database:

1. Define primary keys for each table to ensure uniqueness and that they are NOT NULL

2. Establish foreign keys to enforce referential integrity, ensuring records exist in their respective linked tables

3. Validate that all date fields are in the appropriate date or timestamp format

4. Specify NOT NULL constraints for all mandatory attributes
